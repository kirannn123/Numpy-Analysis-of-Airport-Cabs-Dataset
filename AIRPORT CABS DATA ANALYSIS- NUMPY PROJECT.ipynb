```python
import numpy as np

taxi = np.genfromtxt("nyc_taxis.csv", delimiter=",",skip_header =True)
```

**MEAN SPEED OF THE CAB RIDES**

```python
speed = taxi[:,7]/ (taxi[:,8]/3600)
```

```python
mean_speed = speed.mean()
print("Mean Speed of the Cab Rides: ",mean_speed)
```

**NUMBER OF RIDES TAKEN IN FEBRUARY MONTH**

```python
rides_feb = taxi[taxi[:,1] == 2,1]
print("Number of rides taken in February: ", rides_feb.shape[0])
```

**NUMBER OF RIDES WITH A TIP GREATER THAN $50**

```python
tip = taxi[taxi[:,-3] > 50, -3]
print("Number of Rides with a trip greater than $50: ", tip.shape[0])
```



**NUMBER OF RIDES WHERE DROP WAS JFK AIRPORT**


```python
jfk = taxi[taxi[:,6] == 2,6]                                                       # location code of jfk airport = 2
print("Number of rides where drop was JFK Airport: ",jfk.shape[0])
```
